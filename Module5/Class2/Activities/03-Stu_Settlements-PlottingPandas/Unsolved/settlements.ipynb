```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
settlements_df = pd.read_csv("Resources/union_settlements_1995.csv")
settlements_df.head()
```

```python
# List columns
settlements_df.columns
```

```python
# Get total settlements by union
union_data = settlements_df['UNION'].value_counts()

union_data
```

```python
# Configure plot, figsize, title, and axis labels
fig1 = union_data.plot(kind="bar", color = "red", figsize=(8,6), title = "Major Collective Bargaining Sttlements (1995)", xlabel="Union", ylabel = "Settlements")
# Configure x-tick rotation
xticklabels = union_data.index
fig1.set_xticklabels(xticklabels, rotation=45, rotation_mode="anchor", ha="right", wrap=True)
# Show plot
plt.tight_layout()
plt.show()
```
