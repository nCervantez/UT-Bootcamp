```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1,13,1)
```

```python
# Avearge weather temp
points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
celsius = [(p - 32) * 0.56 for p in points_F]
```

```python
# Create a handle for each plot
f_handle, = plt.plot(x_axis, points_F, marker ='o', color='blue', label="Farenheit")
c_handle, = plt.plot(x_axis, celsius, marker ='^', color='red', label="Celsius")
```

```python
# Set our legend to where the chart thinks is best
f_handle, = plt.plot(x_axis, points_F, marker ='o', color='blue', label="Farenheit")
c_handle, = plt.plot(x_axis, celsius, marker ='^', color='red', label="Celsius")
plt.legend()
```

```python
# Create labels for the X and Y axis
plt.xlabel("Months")
plt.ylabel("Temperature")
f_handle, = plt.plot(x_axis, points_F, marker ='o', color='blue', label="Farenheit")
c_handle, = plt.plot(x_axis, celsius, marker ='^', color='red', label="Celsius")
plt.legend()
plt.show()
```

```python
# Save and display the chart
plt.xlabel("Months")
plt.ylabel("Temperature")
f_handle, = plt.plot(x_axis, points_F, marker ='o', color='blue', label="Fahrenheit")
c_handle, = plt.plot(x_axis, celsius, marker ='^', color='red', label="Celsius")
plt.legend()
plt.savefig("../Images/Temperature.png")
plt.show()
```
