## 1. What are the geocoordinates (latitude and longitude) of Brisbane, Australia?

```python
# Set the search parameters
target_city = "Brisbane, Australia"

# Build URL using the geocode endpoint
target_url = f"https://api.geoapify.com/v1/geocode/search?text={target_city}&format=json&apiKey={geoapify_key}"
```

```python
# Run request
geo_data = requests.get(target_url).json()

# Print the json
pprint(geo_data)
```

```python
# Extract lat/lon
lat = geo_data["results"][0]["lat"]
lon = geo_data["results"][0]["lon"]

# Print results
print(f"{target_city}: {lat},{lon}")
```

## 2. What are the geocoordinates (latitude and longitude) of The Sydney Opera House?

```python
# update params dict
base_url = "https://api.geoapify.com/v2/places"
latitude = -33.85719805
longitude = 151.21512338473752
categories = "entertainment.culture.arts_centre"
radius = 8000
filters = f"circle:{longitude},{latitude},{radius}"
bias = f"proximity:{longitude},{latitude}"
limit = 20




params = {
    "categories":categories,
    "limit":limit,
    "filter":filters,
    "bias":bias,
    "apiKey":geoapify_key    
}

# Build URL using the geocode endpoint
response = requests.get(base_url, params=params)
response
```

```python
# Run request


# Print the json (pretty printed)
places_data = response.json()
pprint(places_data)
```

```python
# Extract lat/lon
lat = places_data["features"][0]["properties"]["lat"]
lon = places_data["features"][0]["properties"]["lon"]

# Print results
print(f"Sydney Opera House: {lat},{lon}")
```
